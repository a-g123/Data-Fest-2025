```python
import pandas as pd

# Load the data
df3 = pd.read_csv("Price and Availability Data.csv")

# Check the first few rows of the dataset
print(df3.head())

# Basic statistics for 'available_space' and 'overall_rent'
print(df3[['available_space', 'overall_rent']].describe())

# Check for any missing values in the columns
print(df3[['available_space', 'overall_rent']].isnull().sum())

# Correlation between available_space and overall_rent
correlation = df3[['available_space', 'overall_rent']].corr()
print(f"Correlation between available_space and overall_rent: \n{correlation}")

# Scatter plot for visual comparison
import matplotlib.pyplot as plt

plt.scatter(df3['available_space'], df3['overall_rent'])
plt.title('Available Space vs. Overall Rent')
plt.xlabel('Available Space')
plt.ylabel('Overall Rent')
plt.show()

# Optional: Regression line to visualize trend (if applicable)
import numpy as np
from sklearn.linear_model import LinearRegression

# Prepare the data for linear regression
X = df3[['available_space']].dropna()
y = df3['overall_rent'].dropna()

# Fit the model
model = LinearRegression()
model.fit(X, y)

# Predicting using the model
predictions = model.predict(X)

# Plot the regression line
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, predictions, color='red', label='Regression line')
plt.title('Linear Regression: Available Space vs. Overall Rent')
plt.xlabel('Available Space')
plt.ylabel('Overall Rent')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df3 = pd.read_csv("Price and Availability Data.csv")

# Step 1: Calculate average rent by market
market_rent = df3.groupby('market')['sublet_overall_rent'].mean().reset_index()

# Step 2: Find the median rent
median_rent = market_rent['sublet_overall_rent'].median()

# Step 3: Label markets as high-rent or low-rent
market_rent['rent_category'] = market_rent['sublet_overall_rent'].apply(
    lambda x: 'High-Rent Market' if x > median_rent else 'Low-Rent Market'
)

# Merge this info back into the original dataframe
df3 = df3.merge(market_rent[['market', 'rent_category']], on='market')

# Step 4: Group by rent category and sum sublet available space
grouped = df3.groupby('rent_category')['sublet_available_space'].sum().reset_index()

# Plot
plt.figure(figsize=(8, 5))
plt.bar(grouped['rent_category'], grouped['sublet_available_space'], color=['skyblue', 'salmon'])
plt.title("Sublet Available Space by Market Rent Category")
plt.ylabel("Total Sublet Available Space (sq ft)")
plt.xlabel("Market Type")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load both datasets
df3 = pd.read_csv("Price and Availability Data.csv")
dfl = pd.read_csv("Leases.csv")

# Aggregate sublet available space by market
sublet_space_by_market = dfl.groupby('market')['sublet_available_space'].sum().reset_index()

# Average unemployment rate by market
unemployment_by_market = df3.groupby('market')['unemployment_rate'].mean().reset_index()

# Merge the two on 'market'
merged_df = pd.merge(sublet_space_by_market, unemployment_by_market, on='market')

# Plotting
plt.figure(figsize=(10, 6))
sns.scatterplot(data=merged_df,
                x='unemployment_rate',
                y='sublet_available_space',
                hue='market',
                palette='tab10',
                s=100)

plt.title('Unemployment Rate vs Sublet Available Space by Market')
plt.xlabel('Average Unemployment Rate')
plt.ylabel('Total Sublet Available Space (sq ft)')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: GDP data
gdp_data = {
    'year': [2018, 2019, 2020, 2021, 2022, 2023],
    'GDP': [20.66, 21.54, 21.35, 23.68, 26.01, 27.72]  # in trillion USD
}
gdp_df = pd.DataFrame(gdp_data)

# Step 2: Load and aggregate sublet availability data
df3 = pd.read_csv("Price and Availability Data.csv")

# Ensure 'year' column exists — extract it from date if needed
if 'year' not in df3.columns:
    df3['year'] = pd.to_datetime(df3['date'], errors='coerce').dt.year

# Aggregate average sublet availability proportion by year
sublet_by_year = df3.groupby('year')['sublet_availability_proportion'].mean().reset_index()

# Step 3: Merge GDP and sublet data on year
merged_df = pd.merge(gdp_df, sublet_by_year, on='year')

# Step 4: Plot
fig, ax1 = plt.subplots(figsize=(10, 6))

# GDP plot
sns.lineplot(data=merged_df, x='year', y='GDP', marker='o', ax=ax1, color='blue', label='GDP (Trillions USD)')
ax1.set_ylabel('GDP (Trillions USD)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Create a second y-axis for sublet proportion
ax2 = ax1.twinx()
sns.lineplot(data=merged_df, x='year', y='sublet_availability_proportion', marker='o', ax=ax2, color='green', label='Sublet Availability Proportion')
ax2.set_ylabel('Sublet Availability Proportion', color='green')
ax2.tick_params(axis='y', labelcolor='green')

plt.title('Comparison of GDP and Sublet Availability Proportion by Year')
plt.tight_layout()
plt.show()

# Step 5: Correlation
correlation = merged_df['GDP'].corr(merged_df['sublet_availability_proportion'])
print(f"Correlation coefficient between GDP and sublet availability proportion: {correlation:.3f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
df2 = pd.read_csv("Unemployment.csv")  # contains unemployment_rate
df3 = pd.read_csv("Price and Availability Data.csv")  # contains sublet_availability_proportion

# Ensure 'year' columns exist
if 'year' not in df2.columns:
    df2['year'] = pd.to_datetime(df2['date'], errors='coerce').dt.year

if 'year' not in df3.columns:
    df3['year'] = pd.to_datetime(df3['date'], errors='coerce').dt.year

# Group by year and average
unemployment_by_year = df2.groupby('year')['unemployment_rate'].mean().reset_index()
sublet_by_year = df3.groupby('year')['sublet_availability_proportion'].mean().reset_index()

# Merge on year
merged = pd.merge(unemployment_by_year, sublet_by_year, on='year')

# Plot
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot unemployment
sns.lineplot(data=merged, x='year', y='unemployment_rate', marker='o', color='red', ax=ax1, label='Unemployment Rate')
ax1.set_ylabel('Unemployment Rate (%)', color='red')
ax1.tick_params(axis='y', labelcolor='red')

# Second axis for sublet proportion
ax2 = ax1.twinx()
sns.lineplot(data=merged, x='year', y='sublet_availability_proportion', marker='o', color='green', ax=ax2, label='Sublet Availability Proportion')
ax2.set_ylabel('Sublet Availability Proportion', color='green')
ax2.tick_params(axis='y', labelcolor='green')

plt.title('Unemployment Rate vs Sublet Availability Proportion by Year')
plt.tight_layout()
plt.show()

# Correlation
correlation = merged['unemployment_rate'].corr(merged['sublet_availability_proportion'])
print(f"Correlation coefficient: {correlation:.3f}")
```
